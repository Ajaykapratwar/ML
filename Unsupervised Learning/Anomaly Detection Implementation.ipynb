```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('healthcare.csv')
```

```python
df.head()
```

```python
plt.scatter(df.iloc[:,0], df.iloc[:,1])
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
clt = IsolationForest(contamination=0.2)
clt.fit(df)
prediction = clt.predict(df)
```

```python
prediction
```

```python
import numpy as np
index = np.where(prediction < 0)
index
```

```python
x = df.values
```

```python
index = np.where(prediction < 0)
plt.scatter(df.iloc[:,0], df.iloc[:,1])
plt.scatter(x[index,0], x[index,1], edgecolors='r')
plt.show()
```
