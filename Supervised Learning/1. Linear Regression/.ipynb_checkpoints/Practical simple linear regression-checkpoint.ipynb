## Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('height-weight.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Weight'], df['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
# correlation
df.corr()
```

```python
## use of seaborn for visualization
sns.pairplot(df)
plt.show()
```

```python
# create separate columns for independent and dependent features
X = df[['Weight']]   ### independent features should be data frame or 2 dimensional array
y = df['Height']   ## this variable can be in series or 1d array
```

```python
X.head()
```

```python
X_series = df['Weight']
np.array(X_series).shape
```

```python
np.array(y).shape
```

```python
## Train test split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# standardization
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
## Applying simple linear regression

from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression(n_jobs=-1)
```

```python
regression.fit(X_train, y_train)
```

```python
print("Coefficient or Slope : ", regression.coef_)
print("Intercept : ", regression.intercept_)
```

```python
## Plot best fit line of training data

plt.scatter(X_train, y_train)
plt.plot(X_train, regression.predict(X_train))
plt.show()
```

## Prediction of Test data
1. Predicted height output = intercept + coef_(Weights)
2. y_pred_test = 156.470 + 17.29(X_test)

```python
## prediction of test data
y_pred = regression.predict(X_test)
```

```python
## Performance metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

## R Square
Formula

R^2 = 1 - SSR/SST

R^2 = coefficient of determination SSR = sum of squares of residuals SST = total sum of squares

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test, y_pred)
print(score)
```

### Adjusted R2 = 1 – [(1-R2)*(n-1)/(n-k-1)]

where:

R2: The R2 of the model n: The number of observations k: The number of predictor variables

```python
# display adjusted r squared
1 - (1 - score) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
```
